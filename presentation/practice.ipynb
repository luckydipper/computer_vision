```python
class FullyConnectedLayer(object):
    def __init__(self,input_size, layer_size, activate_func):
        super().__init__()
        self.W = np.random.standard_normal(size=(input_size,layer_size))

        self.b = np.random.standard_normal(layer_size)
        print(self.b.shape)
        self.num_neuron = layer_size # == output_vector size 
        self.activate_func = activate_func
    def forward(self, x): # 다음 layer로 전달
        output_vector = np.dot(x,self.W) + self.b
        return self.activate_func(output_vector)
        
        
if __name__ == "__main__":
    sample = np.random.standard_normal((3,2))
    sigmoid_func = lambda x: 1 / (1 + np.exp(-x))
    input_layer = FullyConnectedLayer(2, 3, sigmoid_func)
    print(input_layer.forward(sample))
    
    print("hello world")
```

```python
import tensorflow as tf
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train, x_test = x_train/255.0 , x_test /255.0
#normalize를 해줘야지 수렴이 잘 되는 듯
num_classes = 10
model = tf.keras.models.Sequential() # I think pipline 느낌??
layer = tf.keras.layers
model.add(layer.Flatten())
model.add(layer.Dense(128,activation='relu'))
model.add(layer.Dense(num_classes, activation='softmax'))
model.compile(optimizer='sgd', loss='sparse_categorical_crossentropy',metrics=['accuracy'])# metrix로 loss 뿐 아니라 accuracy까지 나옴.
model.fit(x_train,y_train,epochs=5,verbose=1,validation_data=(x_test,y_test))
```

```python
A, B = tf.constant(3.), tf.constant(6.)
X = tf.Variable(20.)
loss = tf.math.abs(A * X -B)
def train_step():
    with tf.GradientTape() as tape:
        loss = tf.math.abs(A * X -B)
    dX = tape.gradient(loss,X)
    print("x = {:2f}, dx = {:2f}".format(X.numpy(),dX))
    X.assign(X - dX)
    
for i in range(7):
    train_step()
```
